```python
import cv2
```

```python
# Read an Image
img = cv2.imread("./itachi.jpg")
```

```python
#To Display Image we Use Matplotlib
import matplotlib.pyplot as plt
```

```python
#This is bgr image
plt.imshow(img)
# to hide <matplotlib.image.AxesImage at 0x2093952c288>
# use plt.show()
```

```python
#This is bgr image
plt.imshow(img)
plt.show()
```

Basically , opencv when it reads the image. The image is rgb image it has 3 channels representing each color component. When we read an image using opencv, so opencv treat the image by default as BGR.R and B got swapped. 

**Open CV treats the image as BGR**

```python
plt.figure(figsize=(20,20))
newimg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(newimg)
plt.axis("off")
plt.show()
```
